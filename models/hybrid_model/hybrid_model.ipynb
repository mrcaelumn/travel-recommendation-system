```python
import numpy as np
import pandas as pd
import sklearn
```

```python
df_train = pd.read_csv('data/final_data_train.csv')
print(df_train.shape[0])
df_train.head()
```

```python
print(df_train.columns)
```

```python
from sklearn.cluster import KMeans
df_train_final = df_train[['user_id', 'business_id','review_star','sentiment_score', 'state', 'city']]
```

```python
from sklearn.preprocessing import LabelEncoder

# encode categorical variables as numerical labels
le = LabelEncoder()
df_train_final['state'] = le.fit_transform(df_train_final['state'])
df_train_final['city'] = le.fit_transform(df_train_final['city'])
```

```python
df_train_final.head()
```

```python
df_train_final.to_csv("data/final_data_reviews_kmeans.csv")
```

```python
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


# scale the data
scaler = MinMaxScaler()
X = scaler.fit_transform(df_train_final)

max_cluster = 20
# calculate the within-cluster sum of squares for different values of k
wcss = []
for k in range(1, max_cluster):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# plot the elbow curve
plt.plot(range(1, max_cluster), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=6, init='k-means++')
kmeans.fit(X)
y = kmeans.labels_
```

```python
print(y)
```

```python
df_train_final['cluster'] = kmeans.predict(df_train_final[['review_star','sentiment_score', 'state', 'city']])
df_train_final.head()
```

```python
import pickle
# save the model weights to a file
with open('weight/kmeans_weights.pkl', 'wb') as f:
    pickle.dump(kmeans, f)
```

```python
df_train_knn = pd.read_csv('data/final_data_train.csv')
```

```python
df_train_knn_final = df_train_knn[['user_id','business_id', 'review_star']]
print(df_train_knn_final.shape[0])
df_train_knn_final.head()
```

```python
df_train_knn_final = df_train_knn_final.groupby(['user_id', 'business_id'])['review_star'].mean().unstack().fillna(0)
```

```python
df_train_knn_final.head()
```

```python
df_train_knn_final.to_csv("data/user_rating_matrix_knn.csv")
```

```python
# get the column names
columns = df_train_knn_final.columns

# create an empty DataFrame with only the columns
empty_df = pd.DataFrame(columns=columns)

# write the column names to a CSV file
empty_df.to_csv('data/final_data_reviews_knn.csv', index=False)
```

```python
from sklearn.neighbors import NearestNeighbors

# fit a KNN model
knn_model = NearestNeighbors(metric='cosine', algorithm='brute')
knn_model.fit(df_train_knn_final)
```

```python
import pickle
# save the model weights to a file
with open('weight/knn_weights.pkl', 'wb') as f:
    pickle.dump(knn_model, f)
```
